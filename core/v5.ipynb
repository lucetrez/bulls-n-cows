```python
%pip install tqdm

import matplotlib.pyplot as plt
from tqdm import tqdm
import numpy as np

from v0.game import BullsNCows
from v0.tools import (
    guess_based_on_score,
    guess_based_on_entropy,
    guess_based_on_candidates,
)
```

```python
import random
digits, n_iter=2, 1000
game = BullsNCows(digits=digits, guess_algorithm=guess_based_on_score, verbose=False)
secrets = [random.choice(game.originals) for _ in range(n_iter)]
```

```python
plt.subplot(2, 2, 1)
game = BullsNCows(digits=digits, guess_algorithm=guess_based_on_score, verbose=False)
scores_1 = [game.reset().play(secret=secrets[i], calc_entropy_score=np.sqrt)[0] for i in tqdm(range(n_iter))]
plt.ylim(0, n_iter*2//3)
plt.title(f'guess_based_on_score ({digits}) - {sum(scores_1)/len(scores_1):.2f}')
plt.hist(scores_1, bins=np.arange(0, 10), align='left', edgecolor='black', linewidth=0.4)

plt.subplot(2, 2, 2)
game = BullsNCows(digits=digits, guess_algorithm=guess_based_on_score, verbose=False)
scores_2 = [game.reset().play(secret=secrets[i], calc_entropy_score=np.sqrt)[0] for i in tqdm(range(n_iter))]
plt.ylim(0, n_iter*2//3)
plt.title(f'guess_based_on_score ({digits}) - {sum(scores_2)/len(scores_2):.2f}')
plt.hist(scores_2, bins=np.arange(0, 10), align='left', edgecolor='black', linewidth=0.4)

plt.subplot(2, 2, 3)
game = BullsNCows(digits=digits, guess_algorithm=guess_based_on_score, verbose=False)
scores_3 = [game.reset().play(secret=secrets[i], calc_entropy_score=np.sqrt)[0] for i in tqdm(range(n_iter))]
plt.ylim(0, n_iter*2//3)
plt.title(f'guess_based_on_score ({digits}) - {sum(scores_3)/len(scores_3):.2f}')
plt.hist(scores_3, bins=np.arange(0, 10), align='left', edgecolor='black', linewidth=0.4)

plt.subplot(2, 2, 4)
game = BullsNCows(digits=digits, guess_algorithm=guess_based_on_score, verbose=False)
scores_4 = [game.reset().play(secret=secrets[i], calc_entropy_score=np.sqrt)[0] for i in tqdm(range(n_iter))]
plt.ylim(0, n_iter*2//3)
plt.title(f'guess_based_on_score ({digits}) - {sum(scores_4)/len(scores_4):.2f}')
plt.hist(scores_4, bins=np.arange(0, 10), align='left', edgecolor='black', linewidth=0.4)

plt.tight_layout()
plt.show()
```
